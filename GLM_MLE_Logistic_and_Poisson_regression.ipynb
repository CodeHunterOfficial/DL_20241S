Программа на Python для вычисления вероятности выпадения одного "орла" из восьми бросков с использованием формулы биномиального распределения вероятностей:

```python
import math

# Заданные значения
n = 8  # общее количество бросков
k = 3  # количество "орлов"
p = 0.5  # вероятность "орла" в каждом броске

# Вычисление биномиального коэффициента
binomial_coefficient = math.comb(n, k)

# Вычисление вероятности
probability = binomial_coefficient * (p ** k) * ((1 - p) ** (n - k))

print("Вероятность выпадения одного 'орла' из восьми бросков:", probability)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Заданные значения
n = 8  # общее количество бросков
p = 0.5  # вероятность "орла" в каждом броске

# Создание списка возможных исходов (от 0 до n)
x = np.arange(0, n+1)

# Вычисление вероятностей для всех возможных исходов
probabilities = stats.binom.pmf(x, n, p)

# Построение графика
plt.bar(x, probabilities, align='center', alpha=0.5)
plt.title('Вероятности для всех возможных исходов')
plt.xlabel('Количество "орлов"')
plt.ylabel('Вероятность')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Создание списка вероятностей от 0 до 1 с шагом 0.01
probabilities = np.arange(0, 1.01, 0.01)

# Расчет правдоподобий для заданной вероятности и наблюдаемых данных
likelihoods = binom.pmf(4, 8, probabilities)

# Построение графика
plt.plot(probabilities, likelihoods, marker='o')
plt.title('Правдоподобия для разных значений вероятности')
plt.xlabel('Вероятность')
plt.ylabel('Правдоподобие')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Создание списка вероятностей от 0 до 1 с шагом 0.01
p_values = np.arange(0, 1.01, 0.01)

# Расчет правдоподобий для каждого значения p
likelihoods = binom.pmf(6, 8, p_values)

# Построение графика
plt.plot(p_values, likelihoods, marker='o')
plt.title('Правдоподобие для разных значений p')
plt.xlabel('Значение p')
plt.ylabel('Правдоподобие')
plt.show()
```
